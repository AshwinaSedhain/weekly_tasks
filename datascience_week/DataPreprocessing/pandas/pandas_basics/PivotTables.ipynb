```python
import pandas as pd

data = {
    'Date': ['2026-01-01', '2026-01-01', '2026-01-02', '2026-01-02', '2026-01-03'],
    'Store': ['North', 'South', 'North', 'South', 'North'],
    'Category': ['Tech', 'Tech', 'Furniture', 'Tech', 'Furniture'],
    'Sales': [1000, 2000, 1500, 500, 1200],
    'Quantity': [2, 4, 3, 1, 2],
    'Profit': [200, 450, 300, 100, 250]
}

df = pd.DataFrame(data)
```

```python
df
```

```python
#A Pivot Table reshapes your data by turning unique values from one column into Rows (Index) and unique values 
#from another column into Columns, then calculates the Values in the middle.
```

```python
# Rows = Store, Columns = Category, Values = Total Sales
view1 = df.pivot_table(index='Store', columns='Category', values='Sales', aggfunc='sum', fill_value=0)
```

```python
# Rows = Date, Columns = Store, Values = Total Profit
view2 = df.pivot_table(index='Date', columns='Store', values='Profit', aggfunc='sum')
```

```python
# We use a dictionary to define different math for different columns
view3 = df.pivot_table(
    index='Store', 
    aggfunc={'Sales': 'sum', 'Quantity': 'sum', 'Profit': 'mean'}
)
```

```python
# Pass a LIST to index to create sub-groups
sub_report = df.pivot_table(
    index=['Store', 'Category'], 
    values='Sales', 
    aggfunc='sum'
)
```

```python
# Defines the "Logic Map"
logic = {
    'Sales': 'sum',      # Total Revenue
    'Quantity': 'mean',   # Average items per basket
    'Profit': 'max'       # The single highest profit deal
}

health_check = df.pivot_table(index='Date', aggfunc=logic)
```
