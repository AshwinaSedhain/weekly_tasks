```python
import numpy as np
import pandas as pd
```

```python
#fillna = COLUMN WISE
#DROP = ROW WISE
#LOC( SELECTION0 = COLUMM
#FILL= column
```

```python
#FINDING THE MISSING DATAS
```

```python
data = {
    'Name': ['Ashwini', 'Ashwinii', 'Ashwiniii', 'Ramesh', np.nan, 'Sita'],
    'Age': [22, np.nan, 24, 25, 21, np.nan],
    'Salary': [60000, 80000, np.nan, 45000, 70000, 55000],
    'City': ['Kathmandu', np.nan, 'Boston', 'Pokhara', np.nan, 'Kathmandu']
}
df=pd.DataFrame(data)
```

```python
df
```

```python
# IF WE WANT TO SEE IF THERE IS NULL VALUE...IN BOOLEAN VALUE( TRUE OR FALSE)
df.isna()
```

```python
#IF WE WANT TO SUM UP 
#TO FIND HOW MANY NULL VALUES ARE THERE IN ROWS
df.isna().sum()
```

```python
#IF WE WANT TO SEE IF EVERY COLUMN CONSIST OF NULL VAUES OR NOT
df.isna().any()
```

```python
# REMOVING THE MISSING DATA
```

```python
#WHEN WE ARE REMOVING THE MISSING DATA IT WORKS ON THE ROW BASIS
df.dropna()
```

```python
#WE USE THRESHHOLS TO HAVE  A CONDIDTION 
# HERE WE CHECK IF ANY ROW HAVE ATLEAST 3  NON NULL VALUE THESN EXCEPT IT REMOVE ALL OTHER
df.dropna(thresh = 4)
```

```python
#FILLING THE MISSING DATA
```

```python
#FOR REPLACING THE MSIIING NUMERIC MISSING VALUE WE REPLACE WITH 0.
# df.fill(0)---its not enought to keep change permanantly so inorder to keep
#the 0 in missing values we have to use of inplace(inplace = true)
```

```python
df.fillna(0)
```

```python
#filling nan with different values
#her ewe have string values where we cant put the numeric 0 for the missing value 
#so its best to write unkonwn in this case
values= {'Name':'unknown','Age':0, 'Salary':0, 'City':'unknown'}
df.fillna(values)
```
