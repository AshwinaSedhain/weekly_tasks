```python
# DATAFRAME
#WHEN YOU COMBINE THE MULTIPLE SERIES THEN ,IT IS CALLED DATAFRAME.
#It consists of multiple series but one indexing.
import numpy as np
import pandas as pd

```

```python
#CREATING THE DATAFRAME
#WE CAN USE LIST,ARRAY,DICT ETC..TO CREATE A DATAFRAME.
#USING DICTONARY
data = {
    'Name':['Ashwini', 'Ashwina', 'Ashmina' ],
    'Age' :[21, 22, 23],
    'city' :['london', 'Boston', 'newyork'],
    'salary':[70000,80000,90000]
}
df = pd.DataFrame(data)
```
