```python
import numpy as np
import pandas as pd
```

```python
df_employees = pd.DataFrame({
    'Emp_ID': [101, 102, 103, 104],
    'Name': ['Ashwini', 'sital', 'Savyata', 'Monika'],
    'Dept': ['Engineering', 'Marketing', 'Engineering', 'HR']
})

df_salaries = pd.DataFrame({
    'Emp_ID': [101, 102, 103, 105],
    'Annual_Salary': [95000, 72000, 88000, 65000],
    'Bonus_Percent': [0.10, 0.05, 0.12, 0.08]
})
```

```python
df_employees
```

```python
df_salaries
```

```python
# ON = IT MENTION, ON WHICH BASIS WE ARE COLUMNS WHICH ARE BOTH AVAILABLE IN DATAFRAMES
# HOW = INNNER 
#It only keeps rows where the "ID" (or name) exists in both DataFrames. If there isn't a match on both sides, the data is discarded.
```

```python
pd.merge(df_employees,df_salaries , on = 'Emp_ID',how = 'inner')
```

```python
# HOW = OUTER
#It keeps every single row from both DataFrames. If a match doesn't exist, pandas doesn't delete the row; instead,
#it fills the missing gaps with NaN (Not a Number).
```

```python
pd.merge(df_employees,df_salaries , on = 'Emp_ID' , how = 'outer')
```

```python
#how='left'
#keeps everything from the first DataFrame (the one on the left) and only brings in data 
#from the second DataFrame where there is a match.
pd.merge(df_employees,df_salaries , on = 'Emp_ID' , how = 'left')
```

```python
# how='right' 
#is the exact opposite. It keeps everything from the second DataFrame and only pulls in 
#info from the first one if it fits.
pd.merge(df_employees,df_salaries , on = 'Emp_ID' , how = 'right')
```

```python
#CONCATINATION OF 2 DATAFRAMES
# WE SHOULD USE ( PD.CONCAT)
# IF WE WANT TO CONCATINATE THE THE DATAFARAMES ON THE BASIS OF ROWS AND COLUM THEN WE SHOULD USE THE AXIS..
#FOR ROW (AXIS= 1)
#FOR COLUMN (AXIS =0)
```

```python
pd.concat([df_employees,df_salaries] , axis = 1)
```

```python
pd.concat([df_employees,df_salaries], axis = 0)
```

```python
#JONINNG OF 2 DATAFRAMES 
#Joining is a specific way to combine two DataFrames based on their indexes rather than columns.
```

```python
students = pd.DataFrame({
    'Names': ['Leo', 'Mia', 'Max'],
    'Grade': ['5th', '6th', '5th']
})
fees = pd.DataFrame({
    'Name': ['Leo', 'Max', 'Sam'],
    'Paid': ['Yes', 'Yes', 'Yes']
})
```

```python
#Inner Join is a "Match-Only" list (it only keeps items found in both),

students.join(fees, how = 'inner')
```

```python
#while Outer Join is a "Master" list (it keeps everything from both, even if they don't match).
fees.join(students, how = 'outer')
```
