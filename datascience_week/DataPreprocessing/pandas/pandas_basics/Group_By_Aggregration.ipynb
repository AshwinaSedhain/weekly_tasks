```python
import numpy as np
import pandas as pd
```

```python
#GROUP BY AGGREGATION
#Group By splits your data into "buckets" based on a category.
```

```python
data = {
    'Date': ['2026-01-01', '2026-01-01', '2026-01-02', '2026-01-02'],
    'Store': ['North', 'South', 'North', 'South'],
    'Category': ['Electronics', 'Furniture', 'Electronics', 'Electronics'],
    'Sales': [1000, 500, 1200, 800],
    'Quantity': [2, 1, 3, 2]
}
df = pd.DataFrame(data)
```

```python
df 
```

```python
#TOTAL MONEY
#Adds all sales together.
df.groupby('Store')['Sales'].sum()
```

```python
# Find the maximum sale for each store
df.groupby('Store')['Sales'].max()
```

```python
# Find the minimum sale for each store
df.groupby('Store')['Sales'].min()
```

```python
# Group by two columns by passing them in a list []
df.groupby(['Store', 'Category'])['Sales'].sum()
```

```python
#.sum()	Total money made.
#.mean()	Average price per sale.
#.count()	Total number of customers/rows.
#.max()	The most expensive item sold.
#.std()	How much the sales prices vary (consistency).
```
