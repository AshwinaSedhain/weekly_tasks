```python
#importing Libraries for visualization
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#loading cdv
df = pd.read_csv("/home/ashwini/Documents/DataPreprocessing/datas/tips.csv")


```

```python
df.head()
```

```python
#visualization of average tip by day

df.groupby('day')['tip'].mean().plot(kind='bar')

plt.title("Average Tip by Day")
plt.xlabel("Day")
plt.ylabel("Average Tip")
plt.show()
```

```python
# visualization using scatter of total bill vs tip
plt.scatter(df['total_bill'], df['tip'])

plt.title("Total Bill vs Tip")
plt.xlabel("Total Bill")
plt.ylabel("Tip")
plt.show()
```

```python
# sing v histogram to visualize distribution of total bill
plt.hist(df['total_bill'], bins=20)

plt.title("Total Bill Distribution")
plt.xlabel("Total Bill")
plt.ylabel("Frequency")
plt.show()
```

```python
# using boxplot for tip distribution
plt.boxplot(df['tip'])

plt.title("Tip Distribution")
plt.show()
```

```python
# pie chart for smoker vs non-smoker count
df['smoker'].value_counts().plot(kind='pie', autopct='%1.1f%%')

plt.title("Smoker Distribution")
plt.ylabel("")
plt.show()
```
